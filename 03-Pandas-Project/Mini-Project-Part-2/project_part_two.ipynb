```python
# Dependencies
import pandas as pd
```

```python
# load CSV
file = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Read with pandas--low_memory required to suppress errors about mixed data types
coders_file = pd.read_csv(file, low_memory=False)
coders_file.head()
```

```python
# Taking only columns 0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111

i = [0, 1, 2, 3, 4, 6, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]
coders_data = pd.DataFrame([coders_file.iloc[:,i] for i in i]).T
coders_data.head()
```

```python
# Changing reading 0 and 1 to No and Yes, respectively
coders_edit = coders_data.replace( 0, "No")
coders_edit = coders_edit.replace( 1, "Yes")
coders_edit["AttendedBootcamp"].value_counts()
```

```python
# Extracting rows for only those who attended a bootcamp
attended_df = coders_edit.loc[coders_edit["AttendedBootcamp"] == "Yes"]
attended_df.head()
```

```python
# Creating DataFrame of the different boot camps that had a significant number of attendees i.e 20 or more attendees

bootcamps = attended_df["BootcampName"].value_counts()
bootcamps_df = pd.DataFrame(bootcamps).rename(columns={"BootcampName":"Popular BootCamp"})
bootcamps_df = bootcamps_df.loc[bootcamps_df["Popular BootCamp"] >= 20]
bootcamps_df.index.rename("BootCamp", inplace=True)
bootcamps_df
```

```python
# Counting how many attendees of each bootcamp would recommend the bootcamp
recom_bootcamps = attended_df["BootcampName"].loc[attended_df["BootcampRecommend"] == "Yes"].value_counts() #
recom_bootcamps_df= pd.DataFrame(recom_bootcamps).rename(columns={"BootcampName":"Recommended BootCamp"})
recom_bootcamps_df.index.rename("BootCamp", inplace=True)
recom_bootcamps_df
```

```python
# Merging the two created data frames on the name of tbe bootcamp
combined_df = pd.merge(bootcamps_df,recom_bootcamps_df, how="outer", on="BootCamp")
combined_df
```

```python
# Calculate percentage of each bootcamp's students who are recommenders
combined_df['Percentage of recommenders'] =  (combined_df["Recommended BootCamp"]/\
                                              (combined_df["Popular BootCamp"] + combined_df["Recommended BootCamp"]))

combined_df = combined_df.loc[(combined_df["Percentage of recommenders"].notnull())].sort_values("Percentage of recommenders", ascending=False)
combined_df["Percentage of recommenders"]=combined_df["Percentage of recommenders"].map("{:,.2%}".format)
combined_df
```

```python
# Export to excel and remove index
combined_df.to_csv('Output/Attendee Recommenders_Analysis_2016-Bootcamp.csv', index=False, header=True)
```
