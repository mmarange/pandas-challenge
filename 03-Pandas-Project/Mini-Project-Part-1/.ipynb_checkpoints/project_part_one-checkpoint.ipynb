```python
# Dependencies
import pandas as pd
import numpy as np
# Read Purchasing File and store into Pandas data frame
```

```python
# loading CSV
file = "Resources/2016-FCC-New-Coders-Survey-Data.csv"
```

```python
# Reading with pandas
coders_file = pd.read_csv(file, low_memory=False)
coders_file.head()
```

```python
# Inspecting all columns
coders_file.dtypes
responders = coders_file["AttendedBootcamp"].value_counts(dropna=False)
responders
```

```python
# Extracting only columns 0, 1, 2, 3, 4, 7, 8, 9, 10,11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111
i = [0, 1, 2, 3, 4, 7, 8, 9, 10, 11, 29, 30, 32, 36, 37, 45, 48, 56, 110, 111]
coders_data = pd.DataFrame([coders_file.iloc[:,i] for i in i]).T
coders_data.head()
```

```python
# Changing "0" to "No" and "1" to "Yes" in response columns
coders_edit = coders_data.replace(0,"No")
coders_edit = coders_edit.replace(1,"Yes")

coders_edit["AttendedBootcamp"].value_counts(dropna=False) # Checking for abnormal entries
```
